```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('Crimes_With_Dates_Cleaned.csv',index_col=0,low_memory=False)
df['date'] = pd.to_datetime(df['Start_Date_Time'])
df['hour'] = df['date'].apply(lambda x:x.hour) 
hour_data = df['hour'].value_counts()
```

```python
hour_dict = dict(hour_data)
hour_dict = sorted(hour_dict.items())
```

```python
plt.rcParams['font.sans-serif'] = ['Microsoft YaHei']
plt.rcParams['axes.unicode_minus']=False
```

```python
x1 = [i[0] for i in hour_dict]
y1 = [i[1] for i in hour_dict]
```

```python
plt.figure(figsize=(12, 9))
plt.plot(x1,y1, linestyle='-',linewidth=5,color='steelblue',) 
plt.xticks(x1) 
plt.ylabel('Number of crimes')
plt.title('24-hour crime data\n',fontsize=25)
plt.show()
```
